```python
from urllib.request import Request, urlopen
import re

alphabet_no_x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's',
                't', 'u', 'v', 'w', 'y', 'z']

player_names = {}

for letter in alphabet_no_x:
    # Gets HTML of page
    req = Request("https://www.basketball-reference.com/players/" + letter +"/", headers={'User-Agent': 'Mozilla/5.0'})
    webpage = str(urlopen(req).read())

    # Finds data we want from the page
    player_name = "data-stat=\"player\" >(?:<strong>)?<a href=\"(\/players\/\w\/[a-zA-Z0-9]+\.html)\">([a-zA-z\.\s\-0-9\']+)<\/a>"
    names = re.findall(player_name, webpage)
    sanity_check = "(data-stat=\"player\")"
    check = re.findall(sanity_check, webpage)
    
    if ((len(names) + 1) != len(check)):
        print("Count mismatch: " + letter)
    else:
        player_names.update(dict(names))
```

```python
import codecs
stats = pd.DataFrame()
player_id = 1

for url in player_names:
    player_req = Request("https://www.basketball-reference.com" + url, headers={'User-Agent': 'Mozilla/5.0'})
    player_page = str(urlopen(player_req).read())

    player_pattern = '<tr id="(playoffs)?[_]?(advanced)?[\w\.]+" class="[\w\s]+" ><th scope="[\w\s]+" class="[\w\s]+" data-stat="season" >(?:<a href=\"\/players\/\w+\/[a-zA-Z0-9]+\/gamelog\/\d+(?:\/aba\/)?\">)([0-9\-]+)(?:<\/a>)(<span class="sr_star"><\/span>)?(<span class="sr_ring"><\/span>)?<\/th><td class="[\w\s]+" data-stat="age" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td><td class="[\w\s]+" data-stat="team_id" >(?:<a href="\/teams\/\w+\/\d+\.html">)?(\w+)(?:<\/a>)?<\/td><td class="[\w\s]+" data-stat="lg_id" >(?:<a href="\/leagues\/[0-9a-zA-Z_]+\.html">)(\w*)(?:<\/a>)<\/td><td class="[\w\s]+" data-stat="pos" >(\w*)<\/td><td class="[\w\s]+" data-stat="g" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td><td class="[\w\s]+" data-stat="(?:gs|mp)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td><td class="[\w\s]+" data-stat="(?:mp_per_g|per)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td><td class="[\w\s]+" data-stat="(?:fg_per_g|ts_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td><td class="[\w\s]+" data-stat="(?:fga_per_g|fg3a_per_fga_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td><td class="[\w\s]+" data-stat="(?:fg_pct|fta_per_fga_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>(?:<td class="[\w\s]+" data-stat="(?:fg3_per_g|orb_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:fg3a_per_g|drb_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:fg3_pct|trb_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:fg2_per_g|ast_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:fg2a_per_g|stl_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:fg2_pct|blk_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:efg_pct|tov_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:ft_per_g|usg_pct)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:fta_per_g|DUMMY)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:ft_pct|ows)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:orb_per_g|dws)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:drb_per_g|ws)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:trb_per_g|ws_per_48)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:ast_per_g|DUMMY)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:stl_per_g|obpm)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:blk_per_g|dbpm)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:tov_per_g|bpm)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="(?:pf_per_g|vorp)" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?(?:<td class="[\w\s]+" data-stat="pts_per_g" >(?:<strong>)?([0-9\.\-]*)(?:<\/strong>)?<\/td>)?'
    stats_re = re.findall(player_pattern, player_page)

    stats_df = pd.DataFrame(stats_re)

    per_game = stats_df[stats_df[1] != 'advanced']
    advanced = stats_df[stats_df[1] == 'advanced']

    per_game = per_game.drop(columns=[1]).rename(columns={0: "Playoffs", 2: "Season", 3: "All Star", 4: "Championship", 5: "Age", 6: "Team", 7: "League", 8: "Position", 9: "Games", 10: "Games Started", 11: "Minutes Per Game", 12: "Field Goals Per Game", 13: "Field Goal Attempts Per Game", 14: "Field Goal Percentage", 15: "3-Points Per Game", 16: "3-Point Attempts Per Game", 17: "3-Point Percentage", 18: "2-Points Per Game", 19: "2-Point Attempts Per Game", 20: "2-Point Percentage", 21: "Effective Field Goal Percentage", 22: "Free Throws Per Game", 23: "Free Throw Attempts Per Game", 24: "Free Throw Percentage", 25: "Offensive Rebounds Per Game", 26: "Defensive Rebounds Per Game", 27: "Total Rebounds Per Game", 28: "Assists Per Game", 29: "Steals Per Game", 30: "Blocks Per Game", 31: "Turnovers Per Game", 32: "Personal Fouls Per Game", 33: "Points Per Game"})
    per_game_season = per_game[per_game["Playoffs"] != "playoffs"].drop(columns=["Playoffs", "Championship"])
    per_game_playoffs = per_game[per_game["Playoffs"] == "playoffs"].drop(columns=["Playoffs", "All Star"])

    per_game_season["All Star"] = per_game_season["All Star"].replace({'': 0, '<span class="sr_star"></span>': 1})

    per_game_playoffs = per_game_playoffs.add_prefix('Playoffs ')
    per_game_playoffs["Playoffs Championship"] = per_game_playoffs["Playoffs Championship"].replace({'': 0, '<span class="sr_ring"></span>': 1})
    per_game = per_game_season.merge(per_game_playoffs, left_on=["Season", "Team"], right_on=["Playoffs Season", "Playoffs Team"], how='outer')
    per_game = per_game.drop(columns=["Playoffs League"])

    advanced = advanced.drop(columns=[1, 3, 4, 5, 7, 8, 9, 23, 28, 33]).rename(columns={0: "Playoffs", 2: "Season", 6: "Team", 10: "Minutes Played", 11: "Player Efficiency Rating", 12: "True Shooting Percentage", 13: "3-Point Attempt Rate", 14: "Free Throw Attempt Rate", 15: "Offensive Rebound Percentage", 16: "Defensive Rebound Percentage", 17: "Total Rebound Percentage", 18: "Assist Percentage", 19: "Steal Percentage", 20: "Block Percentage", 21: "Turnover Percentage", 22: "Usage Percentage", 24: "Offensive Win Shares", 25: "Defensive Win Shares", 26: "Win Shares", 27: "Win Shares Per 48 Minutes", 29: "Offensive Box Plus/Minus", 30: "Defensive Box Plus/Minus", 31: "Box Plus/Minus", 32: "Value over Replacement Player"})
    advanced_season = advanced[advanced["Playoffs"] != "playoffs"].drop(columns=["Playoffs"])
    advanced_playoffs = advanced[advanced["Playoffs"] == "playoffs"].drop(columns=["Playoffs"]).add_prefix('Playoffs ')
    advanced = advanced_season.merge(advanced_playoffs, left_on=["Season", "Team"], right_on=["Playoffs Season", "Playoffs Team"], how='outer').drop(columns=["Playoffs Season", "Playoffs Team"])

    stats_df = per_game.merge(advanced, on=["Season", "Team"], how='outer').apply(pd.to_numeric, errors='ignore')
    stats_df.insert(0, "Player Name", codecs.escape_decode(player_names[url], 'hex')[0].decode('utf8'))
    stats_df.insert(0, "Player ID", player_id)
    
    if (len(stats) == 0):
        stats = stats_df
    else:
        stats = pd.concat([stats, stats_df], ignore_index=True)
        
    player_id = player_id + 1
```

```python
def name_year_award(pattern, page, column_name, single_year=False):
    names_years = re.findall(pattern, page)
    
    names_years_dict = dict()
    for i in names_years:
        names_years_dict[codecs.escape_decode(i[0], 'hex')[0].decode('utf8').replace('–', '-')] = codecs.escape_decode(i[1], 'hex')[0].decode('utf8')
        
    award = []
    for name, year in zip(stats['Player Name'], stats['Season']):
        if single_year:
            if str(year)[5:] and (int(str(year)[5:]) < 30):
                year = '20' + str(year)[5:]
            elif str(year)[5:]:
                year = '19' + str(year)[5:]
            player = names_years_dict.get(year)
            #player = names_years_dict.get(str(year)[:2] + str(year)[5:])
        else:
            player = names_years_dict.get(year)
        if player and player.replace(' ', '').replace('.', '').replace("'", '').lower() == name.replace(' ', '').replace('.', '').replace("'", '').lower():
            award.append(True)
        else:
            award.append(False)

    stats[column_name] = award
    return sum(award)
```

```python
import requests
from bs4 import BeautifulSoup

def format_values(x):
    x = codecs.escape_decode(x, 'hex')[0].decode('utf8')
    x = re.sub(r"[\[|(]\w+[\]|)]|\*|\s+|\.|\'|\^", "", x)
    return x.replace('–', '-').lower()

all_defensive_tbl = BeautifulSoup(all_defensive_page, 'html.parser').find_all("table",{"class":"wikitable"})
all_defensive_df = pd.read_html(str(all_defensive_tbl[1]))[0]
all_defensive_df = all_defensive_df.T.reset_index(drop=True).T
all_defensive_df = all_defensive_df.applymap(format_values).drop(columns={2, 4}).rename(columns={0: 'Season', 1: 'First Team', 3: 'Second Team'})
```

```python
defensive_first_team = []
defensive_second_team = []

for name, season in zip(stats['Player Name'], stats['Season']):
    if str(season) == 'nan':
        defensive_first_team.append(False)
        defensive_second_team.append(False)
        continue
    temp_df = all_defensive_df[all_defensive_df['Season'] == str(season)]
    name = str(name).replace(' ', '').replace('.', '').replace("'", '').lower()
    if name in np.array(temp_df['First Team']):
        defensive_first_team.append(True)
    else:
        defensive_first_team.append(False)            
    if name in np.array(temp_df['Second Team']):
        defensive_second_team.append(True)
    else:
        defensive_second_team.append(False)
```

```python
stats['First All-Defensive Team'] = defensive_first_team
stats['Second All-Defensive Team'] = defensive_second_team
```
