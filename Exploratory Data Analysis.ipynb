```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv('stats.csv')
stats
```

```python
pd.set_option('display.max_columns', None)
```

```python
stats_only_num = stats.drop(columns=["Player ID", "Player Name", "Season", "Team", "League", "Position", "Playoffs Season", "Playoffs Team", "Playoffs Position", "HOF Year Inducted", "Lifetime Achievement Induction", "Last Season Played", "Hall of Fame"]).fillna(0)
stats_only_num
```

```python
# Center data by subtracting mean from each column and divide by standard deviation
stats_only_num_center = (stats_only_num - stats_only_num.mean(axis=0)) / np.std(stats_only_num)
```

```python
# Plot that shows which columns are most important for each principal component
def plot_pc(col_names, vt, k):
    plt.bar(col_names, vt[k, :], alpha=0.7)
    plt.xticks(col_names, rotation=90);
```

```python
svd = np.linalg.svd

U, S, Vt = svd(stats_only_num_center, full_matrices=False)
```

```python
# Shows the captured variance of each principal component
captured_variance = (S**2) / sum(S**2)
captured_variance
```

```python
# Keeping only the first two principal components, we capture 37.49% of the variance in the original data 
captured_variance[0] + captured_variance[1]
```

```python
# Scree Plot (squared singular values and provides a visualization of the captured variance)
plt.plot(S**2);
```

```python
# Compute first two principal components
pcs = U @ np.diag(S)
pc1 = pcs[:, 0]
pc2 = pcs[:, 1]
```

```python
sns.scatterplot(x=pc1, y=pc2)
plt.xlabel('PC1')
plt.ylabel('PC2');
```

```python
z_score_stats = (stats_only_num - stats_only_num.mean()) / stats_only_num.std()
```

```python
z_score_stats['PC1'] = pc1
z_score_stats['PC2'] = pc2
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Shows the difference in Z-Score for certain features split up by the PC1 values
# Almost all features are positive or very close to zero. But the largest difference in z-score 
# occurs for stats that display great individual preformance for a singe season including the playoffs.
# It also includes players that received awards such as MVP, All Star, First All-NBA Team, etc.
pd.DataFrame(z_score_stats[z_score_stats['PC1'] > 20].mean() - z_score_stats[z_score_stats['PC1'] < 0].mean()).drop(['PC1', 'PC2']).sort_values(0).rename(columns={0: 'Z-Score Difference'})
```

```python
# Shows the difference in Z-Score for certain features split up by the PC2 values
# The largest difference in z-score occurs for stats that display great individual as well as
# team preformance for a single playoff run.
pd.DataFrame(z_score_stats[z_score_stats['PC2'] > 5].mean() - z_score_stats[z_score_stats['PC2'] < -5].mean()).drop(['PC1', 'PC2']).sort_values(0).rename(columns={0: 'Z-Score Difference'})
```
